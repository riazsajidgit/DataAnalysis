# Task:01

Choose an appropriate graph to display the change in sales of each category throughout the year. Display each sale inside a separate graph; *Hint: Make use of subplots

```python
# Function to add value on top of bars in a bargraph
def autolabel(rects, ax, color='r', rot=0):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        # '%g' % is done to drop trainling zeros from decimal position
        ax.annotate('%g' % height,
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', color=color, fontsize=13, rotation=rot)
```

```python
fig, ax = plt.subplots(4,3,figsize=(15,15))
plt.tight_layout()
fig.suptitle('Trend of Sales for each Product Category (Furniture, Office Supplies, Technology) from 2014 - 2017', fontsize=18)
fig.subplots_adjust(top=0.95)
#d = {'0':cats_data_2014, 1:cats_data_2015, 2:cats_data_2016, 3:cats_data_2017}
months = ["January","February","March","April","May","June","July","August","September","October","November","December"]
for i in range(0,4):
    for j in range(0,3):
        if i==0:
            d = cats_data_2014
            y = 2014
        elif i==1:
            d = cats_data_2015
            y = 2015
        elif i==2:
            d = cats_data_2016
            y = 2016
        else:
            d = cats_data_2017
            y = 2017
            
        if j==0:
            cat = "Furniture"
            color = 'teal'
        elif j==1:
            cat = "Office Supplies"
            color = 'slateblue'
        else:
            cat = "Technology"
            color = 'c'
        bar = ax[i][j].bar(months,d.loc[cat], color=color)
        ax[i][j].plot(months,d.loc[cat], '4-.r', label=cat+' '+'Sales'+' - '+str(y))
        if i==3:
            ax[i][j].tick_params('x', labelrotation=45)
            ax[i][j].set_xlabel('Month of the Year', fontsize=14)
        else:
            ax[i][j].set_xticks([])
        if j==0:
            ax[i][j].set_ylabel('Sales in USD ($)', fontsize=14)
            
        ax[i][j].legend(loc='upper left', fontsize=13)
        #autolabel(bar,ax[i][j], rot=90)
```
